Removing outliers in income

```python
from scipy import stats

data = data[(np.abs(stats.zscore(data['Income'])) < 3)]     # Remove observation with more than 3 in Standard Desviation
data.reset_index(inplace=True)                          # Reset index
data = data.drop(columns=['index'])  
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
cate = []
for i in data.columns:
    if (data[i].dtypes == "object"):
        cate.append(i)

print(cate)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data['Marital_Status'].value_counts()
```

**Label Encoding**

```python
lbl_encode = LabelEncoder()
for i in cate:
    data[i]=data[[i]].apply(lbl_encode.fit_transform)
```

```python
data1 = data.copy()
```

**Standardization**

```python
scaled_features = StandardScaler().fit_transform(data1.values)
scaled_features_data = pd.DataFrame(scaled_features, index=data1.index, columns=data1.columns)
```

```python
scaled_features_data.head()
```

**Elbow method**

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,11):
 kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
 kmeans.fit(scaled_features_data)
 wcss.append(kmeans.inertia_)
plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The elbow graph indicates that four will be an optimal number of clusters for this data.

---



Although, Elbow method is a good approach to determine the number of clusters, there is another way with which we can find the ideal number of clusters. It's called the silhouette score analysis.

**Silhouette Score**

```python
from sklearn.metrics import silhouette_score 
```

```python
silhouette_scores = []
for i in range(2,10):
    m1=KMeans(n_clusters=i, random_state=42)
    c = m1.fit_predict(scaled_features_data)
    silhouette_scores.append(silhouette_score(scaled_features_data, m1.fit_predict(scaled_features_data))) 
plt.bar(range(2,10), silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('S(i)', fontsize = 20) 
plt.show()
```

 we are using Silhouette score to measure the value of K.

```python
silhouette_scores
```

Getting the maximum value of silhouette score and adding 2 in index as the index starts from 2.

Hence We can go ahead with either 3 or 4 clusters. 

4 is much better because observations seems to be more evenly distributed among the clusters and making stratgey for 4 cluster is much more convinient.

using K-Means Algorithm.

```python
# Training a predicting using K-Means Algorithm.

kmeans=KMeans(n_clusters=3, random_state=42).fit(scaled_features_data)
pred=kmeans.predict(scaled_features_data)


# Appending those cluster value into main dataframe (without standard-scalar)

data['cluster'] = pred + 1
```
